```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1,20 ,200)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('x data for my graph')
plt.ylabel('y data for my graph')
plt.title('This is my title for my graph')
```

```python
x = np.random.rand(40)
y = np.random.rand(40)
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(10,8))
plt.plot(x,y, '--g')
plt.ylabel('x data graph')
plt.xlabel('y data graph')
plt.title('Title for my graph')
plt.grid()
plt.show()
```

```python
colours = np.random.rand(40)
sizes = 400*np.random.rand(40)

plt.scatter(x,y, c=colours, s=sizes, alpha=.4)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('2D Graph')
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
colours = np.random.rand(50)
sizes = 500*np.random.rand(50)

plt.scatter(x,y , s=sizes, marker = 'p', c = colours)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('2D Graph')
```
